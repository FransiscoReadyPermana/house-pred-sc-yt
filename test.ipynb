```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return (float(tokens[0]) + float(tokens[1]))/2
  try:
    return float (x)
  except:
    return None
```

```python
convert_sqft_to_num('1000')
```

```python
convert_sqft_to_num('2100 - 2850')
```

```python
convert_sqft_to_num('4125Perch')
```
